```python
# if koşul yapısı

x = 10

if x > 5:
  print("x, 5' ten büyüktür.")
```

```python
# if - else koşul yapısı

x = 3

if x > 5:
  print("x, 5'ten büyüktür.")
else:
  print("x, 5'ten büyük veya eşittir.")
```

```python
# if - elif - else koşul yapısı

x = 8

if x > 10:
  print("x, 10'dan büyüktür.")
elif x > 5:
  print("x, 5'ten büyük ama 10'dan küçük veya eşittir.")
else:
  print("x,5'ten küçük veya eşittir.")
```

```python
x = 12
y= 8

if x > 10:
  if y > 5:
    print("x, 10'dan büyük ve y, 5'ten büyüktür.")
  else:
    print("x, 10'dan büyük ama y, 5'ten küçük veya eşittir.")

else:
  print("x, 10dan küçük veya eşittir.")
```

```python
# Mantıksal Operatörler ile Koşullar

x = 7
y = 3

# 'and' operatörü ile

if ( x > 5 and y > 2):
  print("x, 5'ten büyük ve y, 2'den büyüktür.")

# 'or' operatörü ile

if (x > 5 or y > 5):
  print("x veya y, 5'ten büyüktür.")

# 'not' operatörü ile

if not (x > 10):
  print("x, 10'dan büyük değildir.")
```

```python
# Örnek 1: Not Sistemi

notu = 85

if (notu >= 90):
  print("Harf notu A.")
elif (notu >= 80):
  print("Harf notu B.")
elif (notu >= 70):
  print("Harf notu C.")
elif (notu >= 60):
  print("Harf notu D.")
else:
  print("Harf notu F.")
```

```python
# Örnek 2: Yaş Kontrolü

yas = 20

if (yas < 18):
  print("resit değilsin.")
elif (yas < 21):
  print("resitsin ama 21'den küçüksün.")
else:
  print("21'den büyüksün.")
```

```python
# Örnek 3: Sayı Pozitif, Negatif veya Sıfır mı?

sayi = int(input("Bir sayı gir: "))

if (sayi > 0):
  print("Sayı pozitiftir.")
elif (sayi < 0):
  print("Sayı negatiftir.")
else:
  print("Sayı 0'a eşittir.")
```
